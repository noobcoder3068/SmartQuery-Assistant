```python
import pickle, faiss
from sentence_transformers import SentenceTransformer

chunk_path = "/content/drive/MyDrive/chunk_conv/rag_chunks.pkl"
index_path = "/content/drive/MyDrive/chunk_conv/rag_index.faiss"

with open(chunk_path, "rb") as f:
    chunks = pickle.load(f)

index = faiss.read_index(index_path)
print("✅ FAISS index and chunks loaded.")
```

```python
query = input("💬 Enter your question: ")

embed_model = SentenceTransformer("all-MiniLM-L6-v2")
query_embedding = embed_model.encode([query])

k = 3
D, I = index.search(query_embedding, k)

retrieved_chunks = [chunks[i] for i in I[0]]
context = "\n\n---\n\n".join(retrieved_chunks)

print("🧩 Retrieved Chunks:\n")
print(context)
```

```python
output_path = "/content/drive/MyDrive/chunk_conv/context.txt"

with open(output_path, "w") as f:
    f.write(f"{context}\n\n---\n\nQuestion:\n{query}\n\nSQL:")
```
